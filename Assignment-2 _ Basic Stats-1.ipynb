# Feature Scaling (Standardization)

Standardization, also known as Z-score normalization, is a data preprocessing technique used in machine learning to transform numerical features so that they have a mean of 0 and a standard deviation of 1.

This is important to prevent features with larger values from dominating the learning process and to ensure that all features contribute equally to the model.

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']] = ss.fit_transform(df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']])
```

```python
df
```

Conclusion -

Improves algorithm performance :
Standardization can help algorithms converge faster and more efficiently, which can save time and resources. It can also help ensure that each feature contributes equally to the learning process, which can lead to more accurate predictions.

Prevents numerical instability :
Standardization can help prevent sensitivity to the scale of input features, which can lead to skewed outcomes.

Enhances data compression and storage :
Standardization can help compress data and store it more efficiently in a data lakehouse, which can improve the performance of advanced analytics.

# Encoding (One-Hot Encoding)

One-hot encoding (OHE) is a machine learning technique that converts categorical data into numerical data for machines to understand. It's useful when data has no relation to each other, or when the order of integers is treated as a significant characteristic by machine learning algorithms.

```python
#Dropping unneccessary columns
df.drop(['Day','City'],axis=1,inplace=True)
```

```python
df
```

```python
df['Brand'].unique()
```

```python
ohe1 = pd.get_dummies(df['Brand'])
```

```python
ohe1
```

```python
#Replace False with 0 and True with 1 in the entire DataFrame.
ohe1.replace({False: 0, True: 1}, inplace=True)
```

```python
ohe1
```

```python
df = pd.concat([df,ohe1],axis=1)
```

```python
df
```

```python
df['Model'].unique()
```

```python
ohe2 = pd.get_dummies(df['Model'])
```

```python
ohe2
```

```python
#Replace False with 0 and True with 1 in the entire DataFrame.
ohe2.replace({False: 0, True: 1}, inplace=True)
```

```python
df = pd.concat([df,ohe2],axis=1)
```

```python
df
```

Conclusion -

One-hot encoding Prevents algorithms from treating categorical values as numbers: This can lead to incorrect assumptions.

Ensures equal distance between categories: This helps algorithms learn more effectively.

Delivers more complex predictions than single labels
