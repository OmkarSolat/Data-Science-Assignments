# **Assignment : EXPLORATORY DATA ANALYSIS (EDA-1)**


 To conduct a thorough exploratory analysis of the "cardiographic.csv" dataset to uncover insights, identify patterns, and understand the dataset's underlying structure using statistical summaries, visualizations, and data manipulation techniques to explore the dataset comprehensively.

```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Dataset
df = pd.read_csv("/content/Cardiotocographic.csv")
```

```python
df
```

# 1. Data Cleaning/Preparation

```python
df.info()
```

```python
df.shape
```

```python
#Check for missing values
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
#checking for outliers
figure =plt.figure(figsize=(20,15))
sns.boxplot(df)
```

'ALTV' & 'MLTV' columns have Greater number of outliers

```python
df.head()
```

```python
#capping
#IQR handling for ALTV

q1 = np.percentile(df['ALTV'],25)
q3 = np.percentile(df['ALTV'],75)

iqr = q3-q1

lb = q1-1.5*iqr   #lower bound
ub = q3+1.5*iqr   #upper bound
```

```python
df[df['ALTV']>ub]
```

```python
#capping
df['ALTV'][df['ALTV']>ub] = ub
df['ALTV'][df['ALTV']<lb] = lb
```

```python
sns.boxplot(df['ALTV'])
```

```python
#capping
#IQR handling for MLTV

q1 = np.percentile(df['MLTV'],25)
q3 = np.percentile(df['MLTV'],75)

iqr = q3-q1

lb = q1-1.5*iqr   #lower bound
ub = q3+1.5*iqr   #upper bound
```

```python
df[df['MLTV']>ub]
```

```python
#capping
df['MLTV'][df['MLTV']>ub] = ub
df['MLTV'][df['MLTV']<lb] = lb
```

```python
sns.boxplot(df['ALTV'])
```

# 3. Data Visualization

```python
df.head()
```

```python
#histogram
df.hist(figsize=(20,15))
plt.show()
```

```python
# Explore the relationship between two variables - LB & Tendency
plt.scatter(df['LB'], df['Tendency'])
plt.xlabel('LB')
plt.ylabel('Tendency')
plt.title('Baseline Fetal Heart Rate (LB) vs Tendency')
plt.show()
```

```python
#pairplot
sns.pairplot(df)
```
